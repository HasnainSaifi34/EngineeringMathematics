## Kmeans Clustering using sklearn



-----

```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
from sklearn.cluster import KMeans
from PIL import Image
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
image_path = './image.jpg'  
image = Image.open(image_path)  
image = np.array(image) / 255.0  


image.shape
```

```python
# Reshape the image to a 2D array of pixels
pixels = image.reshape(-1, 3)

# Apply K-means clustering
n_clusters = 5  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(pixels)
labels = kmeans.predict(pixels)
segmented_image = labels.reshape(image.shape[:2])

# Map the cluster labels to the RGB values of the cluster centers
segmented_image_rgb = kmeans.cluster_centers_[segmented_image]

# Visualize the original and segmented images
fig, ax = plt.subplots(1, 2, figsize=(15, 10))

ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(segmented_image_rgb)
ax[1].set_title('Segmented Image')
ax[1].axis('off')

plt.show()
```
