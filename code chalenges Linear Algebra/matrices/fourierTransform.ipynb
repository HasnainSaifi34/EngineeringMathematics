### Understanding the DFT

The Discrete Fourier Transform (DFT) is a mathematical technique used to transform a sequence of complex numbers into another sequence, representing the original sequence in the frequency domain. This is particularly useful in signal processing, image analysis, and many other fields.

#### Basic Concept

Given a sequence of $ n $ complex numbers $ x = [x_0, x_1, x_2, \ldots, x_{n-1}] $, the DFT of $ x $ is another sequence of $ n $ complex numbers $ X = [X_0, X_1, X_2, \ldots, X_{n-1}] $. The DFT is defined by the formula:
$
X_k = \sum_{j=0}^{n-1} x_j \cdot \exp\left(-2\pi i \frac{jk}{n}\right)
$
for $ k = 0, 1, 2, \ldots, n-1 $. Here, $ \exp\left(-2\pi i \frac{jk}{n}\right) $ is a complex exponential, representing a rotation in the complex plane.

### DFT Matrix Representation

To efficiently compute the DFT, we can represent it using matrix multiplication. This involves constructing the DFT matrix $ F $, which allows us to compute the DFT as a single matrix-vector multiplication.

#### DFT Matrix $ F $

The DFT matrix $ F $ is an $ n \times n $ matrix, where each element $ F_{jk} $ is defined as:
$
F_{jk} = \exp\left(-2\pi i \frac{jk}{n}\right)
$
for $ j, k = 0, 1, 2, \ldots, n-1 $.

In this matrix:
- $ j $ represents the row index.
- $ k $ represents the column index.
- $ \exp\left(-2\pi i \frac{jk}{n}\right) $ is the complex exponential term.

### Computing the DFT Using the Matrix

Once we have the DFT matrix $ F $, we can compute the DFT of the vector $ x $ by simply multiplying $ F $ by $ x $:
$
X = F \cdot x
$
Here, $ X $ is the resulting vector in the frequency domain.

### Properties of the DFT Matrix

1. **Orthogonality**: The rows and columns of the DFT matrix are orthogonal to each other.
2. **Symmetry**: The DFT matrix exhibits certain symmetry properties that are useful in signal processing.
3. **Inverse DFT**: The inverse of the DFT matrix $ F $ (denoted as $ F^{-1} $) allows us to transform back from the frequency domain to the time domain. It is given by:
   $
   F^{-1} = \frac{1}{n} \cdot \overline{F}
   $
   where $ \overline{F} $ is the conjugate transpose of $ F $.

### Example

Let's consider a simple example with $ n = 4 $. The DFT matrix $ F $ for $ n = 4 $ is constructed as follows:

$
F = \begin{bmatrix}
1 & 1 & 1 & 1 \\
1 & w & w^2 & w^3 \\
1 & w^2 & w^4 & w^6 \\
1 & w^3 & w^6 & w^9 \\
\end{bmatrix}
$
where $ w = \exp\left(-2\pi i / 4\right) = i $ (since $ \exp\left(-2\pi i / 4\right) = \cos(\pi/2) + i\sin(\pi/2) = i $).

So,
$
w^0 = 1, \quad w^1 = i, \quad w^2 = -1, \quad w^3 = -i
$

Plugging these values into the matrix:

$
F = \begin{bmatrix}
1 & 1 & 1 & 1 \\
1 & i & -1 & -i \\
1 & -1 & 1 & -1 \\
1 & -i & -1 & i \\
\end{bmatrix}
$

To transform a vector $ x $ using this matrix, we multiply $ F $ by $ x $:

$
X = F \cdot x
$

This results in the vector $ X $ in the frequency domain, where each element represents the frequency component of the original vector $ x $.

### Summary

- The DFT transforms a sequence of complex numbers into another sequence representing its frequency components.
- The DFT matrix $ F $ allows us to perform this transformation using matrix multiplication.
- The elements of $ F $ are complex exponentials representing rotations in the complex plane.
- By multiplying $ F $ with a vector $ x $, we obtain the DFT of $ x $.

```python
import numpy as np;
import matplotlib.pyplot as plt
```

```python
i = (-1)**0.5
pi=np.pi

n = 4
x = 4*np.random.randn(4)+1

def EmptyMat(m,n):
   Mat =[]
   for i in range(m):
       # i need "n" elements in each column of the matrix
       Mat.append([])
   for i in range(m):
       for j in range(n):
           Mat[i].append(0)
        

   return Mat     

def DFT(n):
    Mat = np.array(EmptyMat(n,n),dtype="complex")
    w = np.exp(-2 * pi *i /n)

    for j in range(n):
        for k in range(n):
            m = j*k;
            value = w**m;
            Mat[j][k]=value;
    
    return Mat



F = DFT(n)


x = np.random.randn(n)


X1 = F@x
X2 = np.fft.fft(x)

print(f'X1 == X2: {np.allclose(X1, X2)}')
print(F) 
# plot them
plt.figure()
plt.plot(abs(X1))
plt.plot(abs(X2), marker='o', color='r', linestyle='None')
plt.plot(F.real[:,n])
plt.show()

```
