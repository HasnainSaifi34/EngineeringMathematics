# Sampling of data (xy coordinates) of a circle

----


```python
def Theta():
 return np.random.randn()*2*np.pi


radius = np.round(np.random.randn()*5)

x_sample = []
y_sample = []
no_of_sample = 100;

for i in range(no_of_sample):
   angle = Theta()
   x_value = radius*(np.cos(angle))
   y_value = radius*(np.sin(angle))

   x_sample.append(x_value)
   y_sample.append(y_value)

# i need to create a matrix with column vectors as x , y coordinate
Circle_Mat = []
for i in range(len(x_sample)):
  Circle_Mat.append([x_sample[i],y_sample[i]])

xDomain = [-6,+6]
plt.scatter(x_sample,y_sample)
plt.xlim(xDomain[0],xDomain[1])
plt.ylim(xDomain[0],xDomain[1])   
plt.show()

```
