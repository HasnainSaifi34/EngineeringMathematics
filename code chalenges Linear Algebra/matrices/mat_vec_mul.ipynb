**Matrix-Vector Multiplication as Weighted Combinations**

In linear algebra, matrix-vector multiplication is a fundamental operation that combines a matrix (`A`) and a vector (`v`) to produce a new vector (`b`). Here's how it can be understood as a weighted combination:

**1. Weighted Combination of Columns in A:**

- Imagine each column of matrix `A` (which has dimensions `m x n`) as a separate vector with `m` elements.
- The vector `v` (with dimensions `n x 1`, typically a column vector) acts as a set of weights.
- To find an element `b_i` (i-th element) of the resulting vector `b` (with dimensions `m x 1`), we perform the following:
    - Take the dot product (a scalar value) of each row of `A` with the corresponding element in `v`.
    - Sum these dot products.

Mathematically, this can be expressed as:

```
b_i = [A_i1, A_i2, ..., A_in] * [v1, v2, ..., vn] (dot product)
   = A_i1 * v1 + A_i2 * v2 + ... + A_in * vn
```

where:
  - `A_i` is the i-th row of `A`.
  - `v_j` is the j-th element of `v`.

Essentially, we're creating a new vector `b` by taking linear combinations of the columns of `A`, weighted by the corresponding elements in `v`.

**2. Weighted Combination of Rows in A (Alternative Viewpoint)**

- While less common, matrix-vector multiplication can also be viewed as a weighted combination of the rows of `A`.
- In this interpretation, each row of `A` is multiplied by the corresponding element in `v`, and the resulting products are summed element-wise across each column to form the new vector `b`.

However, it's important to note that this viewpoint is not as straightforward as the column-based interpretation and can be less intuitive for beginners. The column-based approach is generally considered the primary way to understand matrix-vector multiplication.

**Key Points:**

- Matrix-vector multiplication is defined only when the number of columns in `A` (n) is equal to the number of elements (rows) in `v`.
- The resulting vector `b` has the same number of rows (`m`) as `A`.

**Example:**

Consider the following matrix `A` and vector `v`:

```
A = [[1, 2], [3, 4], [5, 6]]
v = [2, 1]
```

To calculate the first element of `b` (using the column-based interpretation):

```
b_1 = [1, 2] * [2, 1] = (1 * 2) + (2 * 1) = 4
```

Following this process for all rows of `A`, we get the resulting vector `b`:

```
b = [4, 14, 22]
```

````python
v1 = np.array(5*np.random.randn(2))

def GetRotation(theta):
    rotation = np.array([
        [np.cos(theta),-np.sin(theta)],
        [np.sin(theta),np.cos(theta)]
    ])
    return rotation;
pi = np.pi;
angles = [0,pi/2,pi,3*pi/2,2*pi ]

labels = ['v1','v2','v3','v4','v5']
vectors = []
for i in range(len(angles)):
    v = GetRotation(angles[i]) @ v1;
    vectors.append(v)

plot_vectors(vectors,['r','g','b','r','g'],labels)    
````

````python
# impure Rotation Rotates as well as change the magnitude of the vector 
v1 = (5*np.random.randn(2))
impureRotation = 4 * GetRotation(pi/2)
v2 = impureRotation @ v1
plot_vectors([v1,v2],['r','b'],['v1','impure rotation of v1'])
print("v2 impure rotation of v1 :",v2)
print("v1 :",v1)
````
