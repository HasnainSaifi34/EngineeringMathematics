```python
# implementation of layering in python NOTE: this code will work with only square matrices

m = 6;
n = 6;
k = 6;

A = np.round(5*np.random.randn(m,n))
B = np.round(5*np.random.randn(n,k))

print("matrix A :")
print(A)
print("\n")
print("matrix B :")
print(B)
print("\n")

 # this will store all layers regarding the layering addition

# loop to extract features or layers
def ExtractLayers(A,B):
  layers = []
  if A.shape[1] != B.shape[0]:
    raise ValueError("Incompatible shapes for outer product. A.shape[1] (", A.shape[1], ") must equal B.shape[0] (", B.shape[0], ")")

  for i in range(min(A.shape[0], B.shape[1])):
    M = np.array(np.outer(A[:,i],B[i,:]))
    layers.append(M)

  for i in range(len(layers)):
    print(f"layer no {i+1}")
    print(layers[i])
    print("\n") 
  return layers


layers = ExtractLayers(A,B)    
sum_matrix = 0;
for layer in layers:
  sum_matrix+=layer;

print("matrix sum(layers) :")
print(sum_matrix)
print("\n")

print("A*B :")
print(np.matmul(A,B))

  

```
