The concepts of column space, row space, null space, and left null space are fundamental in linear algebra. Let's delve into each of these spaces and their orthogonality relationships in detail.

### Definitions and Relationships

1. **Column Space (C(A))**:
   - The column space of a matrix $ A $, denoted as $ \text{Col}(A) $, is the set of all possible linear combinations of the columns of $ A $.
   - It represents all the vectors that can be reached by multiplying $ A $ with some vector $ \mathbf{x} $.

2. **Row Space (R(A))**:
   - The row space of a matrix $ A $, denoted as $ \text{Row}(A) $, is the set of all possible linear combinations of the rows of $ A $.
   - It represents all the vectors that can be obtained as the transpose of $ A $ multiplied by some vector.

3. **Null Space (N(A))**:
   - The null space of a matrix $ A $, denoted as $ \text{Null}(A) $, is the set of all vectors $ \mathbf{x} $ such that $ A \mathbf{x} = \mathbf{0} $.
   - It consists of all solutions to the homogeneous equation $ A \mathbf{x} = \mathbf{0} $.

4. **Left Null Space (N(A^T))**:
   - The left null space of a matrix $ A $, denoted as $ \text{Null}(A^T) $, is the set of all vectors $ \mathbf{y} $ such that $ A^T \mathbf{y} = \mathbf{0} $.
   - It consists of all solutions to the homogeneous equation $ A^T \mathbf{y} = \mathbf{0} $.

### Orthogonality

Orthogonality refers to the concept that two subspaces are orthogonal if every vector in one subspace is orthogonal to every vector in the other subspace. Here’s how these spaces are orthogonally related:

1. **Row Space and Null Space**:
   - The row space of $ A $ is orthogonal to the null space of $ A $.
   - This means that any vector in the row space is orthogonal to any vector in the null space.
   - Mathematically, if $ \mathbf{r} $ is a row vector of $ A $ and $ \mathbf{v} $ is a vector in $ \text{Null}(A) $, then $ \mathbf{r} \cdot \mathbf{v} = 0 $.

2. **Column Space and Left Null Space**:
   - The column space of $ A $ is orthogonal to the left null space of $ A $.
   - This means that any vector in the column space is orthogonal to any vector in the left null space.
   - Mathematically, if $ \mathbf{c} $ is a column vector of $ A $ and $ \mathbf{u} $ is a vector in $ \text{Null}(A^T) $, then $ \mathbf{u}^T \cdot \mathbf{c} = 0 $.

### Orthogonality of Subspaces

1. **Column Space and Left Null Space**:
   $
   \text{C}(A) \perp \text{N}(A^T)
   $
   This means that any vector in the column space of $A$ is orthogonal to any vector in the left null space of $A$. For a vector $\mathbf{v}$ in the column space and $\mathbf{u}$ in the left null space:
   $
   \mathbf{v} \cdot \mathbf{u} = 0
   $
   **Reason**: If $\mathbf{v} \in \text{C}(A)$, then $\mathbf{v} = A \mathbf{x}$ for some vector $\mathbf{x}$. If $\mathbf{u} \in \text{N}(A^T)$, then $A^T \mathbf{u} = \mathbf{0}$. The dot product is:
   $
   \mathbf{v} \cdot \mathbf{u} = (A \mathbf{x}) \cdot \mathbf{u} = \mathbf{x}^T (A^T \mathbf{u}) = \mathbf{x}^T \mathbf{0} = 0
   $
   Hence, $\text{C}(A)$ is perpendicular to $\text{N}(A^T)$.

2. **Row Space and Null Space**:
   $
   \text{R}(A) \perp \text{N}(A)
   $
   This means that any vector in the row space of $A$ is orthogonal to any vector in the null space of $A$. For a vector $\mathbf{v}$ in the row space and $\mathbf{u}$ in the null space:
   $
   \mathbf{v} \cdot \mathbf{u} = 0
   $
   **Reason**: If $\mathbf{v} \in \text{R}(A)$, then $\mathbf{v}$ is a linear combination of the rows of $A$. If $\mathbf{u} \in \text{N}(A)$, then $A \mathbf{u} = \mathbf{0}$. The dot product is:
   $
   \mathbf{v} \cdot \mathbf{u} = \mathbf{v}^T \mathbf{u} = (A \mathbf{x})^T \mathbf{u} = \mathbf{x}^T A^T \mathbf{u} = \mathbf{x}^T \mathbf{0} = 0
   $
   Hence, $\text{R}(A)$ is perpendicular to $\text{N}(A)$.

### Dimensions and Orthogonality Relations

1. **For Column Space and Left Null Space**:
   $
   \text{dim}(\text{C}(A)) + \text{dim}(\text{N}(A^T)) = m
   $
   where $m$ is the number of rows of $A$.

2. **For Row Space and Null Space**:
   $
   \text{dim}(\text{R}(A)) + \text{dim}(\text{N}(A)) = n
   $
   where $n$ is the number of columns of $A$.

These relationships are a reflection of the orthogonality and completeness of the subspaces in $\mathbb{R}^m$ and $\mathbb{R}^n$.

```python

# Define the matrix A dim(C(A)) = 2 dim(R(A)) = 2 
A = np.array([
    [0,    2,    0],
    [4,    5,    4],
    [2,    2.5,  2]
])

# Compute the rank of the matrix
rank = lambda A : np.linalg.matrix_rank(A)
print("Rank of A:", rank(A))

# Prepare the plot
fig = plt.figure(figsize=(24, 8))
ax = fig.add_subplot(111, projection='3d')

# Create a mesh grid for plotting the planes
xx, yy = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))

# Column space plane
cp1 = np.cross(A[:, 0], A[:, 1])
z1 = (-cp1[0] * xx - cp1[1] * yy) / cp1[2]
ax.plot_surface(xx, yy, z1, alpha=0.5, color='blue', label="Column Space")

# Row space plane
rp1 = np.cross(A[0, :], A[1, :])
z2 = (-rp1[0] * xx - rp1[1] * yy) / rp1[2]
ax.plot_surface(xx, yy, z2, alpha=0.5, color='green', label="Row Space")

# Plot the column vectors
ax.plot([0, A[0, 0]], [0, A[1, 0]], [0, A[2, 0]], 'k', label="Column 1")
ax.plot([0, A[0, 1]], [0, A[1, 1]], [0, A[2, 1]], 'k', label="Column 2")

# Plot the row vectors
ax.plot([0, A[0, 0]], [0, A[0, 1]], [0, A[0, 2]], 'r-', label="Row 1")
ax.plot([0, A[1, 0]], [0, A[1, 1]], [0, A[1, 2]], 'r-', label="Row 2")

# Customize the plot
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()

# Adjust the view angle for better visualization
ax.view_init(elev=190, azim=6)
```
