## Cauchy-Schwarz Inequality: Summation and Vector Form

The Cauchy-Schwarz Inequality is a fundamental inequality that relates the product of corresponding terms in two sequences (or the dot product of vectors) to the product of their norms (lengths). It holds true for both sequences of real numbers and vectors in an inner product space.

### Summation Form

For any two real number sequences `a_1, a_2, ..., a_n` and `b_1, b_2, ..., b_n`, the Cauchy-Schwarz Inequality states:

$$ \left( \sum_{i=1}^n a_i b_i \right)^2 \leq \left( \sum_{i=1}^n a_i^2 \right) \left( \sum_{i=1}^n b_i^2 \right) $$

This can be interpreted as:

* The square of the sum of the products of corresponding terms (`a_i * b_i`) is always less than or equal to the product of the sums of the squares of the terms (`a_i^2` and `b_i^2`).

### Vector Form

In the context of inner product spaces (like Euclidean space), the Cauchy-Schwarz Inequality relates the dot product (inner product) of two vectors `u` and `v` to their lengths (`||u||` and `||v||`). The dot product is denoted by `<u, v>`.

The inequality can be written as:

$$ <u, v>^2 \leq ||u||^2 ||v||^2 $$

This translates geometrically to:

* The square of the projection of one vector (`u`) onto another (`v`) is always less than or equal to the product of the squares of their lengths.

**Here's a breakdown of the notation:**

* `<u, v>`: Dot product (inner product) of vectors `u` and `v`.
* `||u||`: Norm (length) of vector `u`.

### Equality Condition

Equality in the Cauchy-Schwarz Inequality occurs if and only if one vector is a scalar multiple of the other. In other words, if there exists a constant `k` such that `u = k * v` (or `a_i = kb_i` for all `i` in the summation form), then equality holds.

Understanding both the summation form and vector form of the Cauchy-Schwarz Inequality provides a broader understanding of its application across various mathematical contexts.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

----
<h2> we will demonstrate if cauchys inequality holds for any given vector </h2>
<h2> <b> | u . v | < = ||u||-||v||</b></h2>


----

```python
v1 = np.array(np.random.randn(3))
v2 = np.array(np.random.randn(3))

dp = np.dot(v1,v2)

NormV1 = np.linalg.norm(v1)
NormV2 = np.linalg.norm(v2)
NormMultiplication = NormV1*NormV2
ang = np.arccos(dp/NormMultiplication)
print("Angle Between Vectors :",ang)
print(f"Dp :{dp} NormMultiplication:{NormMultiplication}")
if(abs(dp)<NormMultiplication):
    print("the inequality holds")
elif(dp==NormMultiplication):
    print("the angle between vectors is zero :",ang)
     
# Create a 3D plot

def Plot3d(v1,v2,label1,label2): 
  fig = plt.figure()
  ax = fig.add_subplot(111, projection='3d')

  # Plot the origin
  ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label=f'{label1}')
  ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label=f'{label2}')

  # Set labels for axes
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')

 # Set the limits of the plot
  ax.set_xlim([-3, 3])
  ax.set_ylim([-3, 3])
  ax.set_zlim([-3, 3])

 # Add a legend
  ax.legend()

# Show the plot
  plt.show()

Plot3d(v1,v2,"v1","v2")
```

<h2>if w is a non zero vector such that u = aw angle between u , w can either be (0,180) or (0, pi) w , u are linearly dependent</h2>

<h3><b> there for | u . v | == |u|*|v| </b> </h3>


----

```python

# Generate random vector u
u = np.random.randn(3)
# Generate a random scalar
random_number = np.random.uniform(low=0.0, high=10.0)
# Generate vector w, which is a scalar multiple of u
w = u * random_number
# Calculate the dot product of u and w
uw = np.dot(u, w)
# Calculate the norms of u and w
NormU = np.linalg.norm(u)
NormW = np.linalg.norm(w)
# Calculate the product of magnitudes of u and w
UW = NormU * NormW
# Ensure numerical precision
# Calculate the angle between the two vectors
angUW = np.arccos(uw / UW)
# Print the angle in degrees and radians
print("Angle between the two vectors (degrees):", np.degrees(angUW))
print("Angle between the two vectors (radians):", angUW)
# Verify the Cauchy-Schwarz inequality
if abs(uw) < UW:
    print("The Cauchy-Schwarz inequality holds.")
elif abs(uw) == UW:
    print("The angle between vectors is zero or pi (0 or 180 degrees).")

Plot3d(u,w,"u","w")
```
