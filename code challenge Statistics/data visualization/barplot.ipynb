```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("./data/countries.csv")
df = df.dropna()

RankMatrix = np.array(
    [
        df["2015 Rank"],
        df["2014 Rank"],
        df["2013 Rank"]
    ]
)

ScoreMatrix = np.array(
    [
        df["2015 Score"],
        df["2014 Score"],
        df["2013 Score"]
    ]
)


CountryName = df["Country Name"]

sample = 6

# Prepare data for sampled countries
Sampled_Country_Name = CountryName[:sample]  # Select first 'sample' countries
Sampled_Score_Matrix = ScoreMatrix[:, :sample]  # Select corresponding scores for all years

# Define positions for bars on the x-axis (centered)
country_positions = np.arange(len(Sampled_Country_Name))

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as desired

# Create bars for each year
plt.bar(country_positions - 0.2, Sampled_Score_Matrix[0], width=0.4, label="2015")
plt.bar(country_positions, Sampled_Score_Matrix[1], width=0.4, label="2014")
plt.bar(country_positions + 0.2, Sampled_Score_Matrix[2], width=0.4, label="2013")

# Set labels and title
plt.xticks(country_positions, Sampled_Country_Name, rotation=45)  # Rotate x-axis labels for readability
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.title("Happiness Score of Sampled Countries (2013-2015)")

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```

## Working with Random Data 


----

```python

# m = round(abs(int(5*np.random.randn())))+2;
# n = round(abs(int(5*np.random.randn())))+2;

m = 30; # 30 groups
n = 6;  # 6 properties

data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * ((2*i)/(n-1)-1)**2 +(i+1)**2;


```

```python
# show the bars !;

fig , ax = plt.subplots(1,3,figsize=(10,4))

ax[0].bar(range(n),np.mean(data,axis=0))
ax[0].set_title('Bar-Plot')


ax[1].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')
ax[1].set_title('Error Bar')


ax[2].bar(range(n),np.mean(data,axis=0))
ax[2].errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')

ax[2].set_title('Both')

# x-axis spacing starts from 0 - no_of_bars - 1
```

```python
## how to change x-axis spacing

xcrossing = [1,2,4,5,6,9]

plt.bar(xcrossing,np.mean(data,axis=0))
plt.errorbar(range(n),np.mean(data,axis=0),np.std(data,axis=0,ddof=1),marker='s',linestyle='')
plt.show()
```

<h2>Understanding Data as  Matrices</h2>

The idea that "data are groups (rows) x property (columns)" essentially means that data can be represented as a matrix, where each row represents a distinct group or entity, and each column represents a specific property or attribute of those groups. Here's an expanded explanation:

### Understanding Data as Matrices

#### 1. Structure of Data as Matrices
- **Rows (Groups)**: Each row corresponds to an individual group, entity, or observation. For example, in a dataset of students, each row could represent a different student.
- **Columns (Properties)**: Each column corresponds to a specific property, attribute, or feature of the groups. In the student dataset, columns could include properties such as name, age, grade, and major.

#### 2. Example
Consider a dataset where you have information about students:
| Student ID | Name  | Age | Grade | Major       |
|------------|-------|-----|-------|-------------|
| 1          | Alice | 20  | A     | Mathematics |
| 2          | Bob   | 21  | B     | Physics     |
| 3          | Carol | 22  | A     | Chemistry   |

In this dataset:
- Each row (1, 2, 3) represents a different student.
- Each column (Student ID, Name, Age, Grade, Major) represents a different property of the students.

#### 3. Matrix Representation
The data can be represented as a matrix $ D $ where each element $ d_{ij} $ represents the value of the $ j $-th property for the $ i $-th group. For our example:
$ D = \begin{bmatrix}
1 & \text{Alice} & 20 & A & \text{Mathematics} \\
2 & \text{Bob} & 21 & B & \text{Physics} \\
3 & \text{Carol} & 22 & A & \text{Chemistry}
\end{bmatrix} $

### Applications and Advantages

#### 1. Data Analysis
- **Descriptive Statistics**: Matrices allow easy calculation of statistics such as mean, median, variance, and standard deviation for each property.
- **Data Manipulation**: Operations like sorting, filtering, and aggregation can be efficiently performed on matrix structures.

#### 2. Machine Learning
- **Feature Representation**: In machine learning, each row can represent a data sample, and each column can represent a feature, making it convenient for algorithms to process the data.
- **Matrix Operations**: Many machine learning algorithms leverage matrix operations for efficiency. For example, linear regression can be solved using matrix multiplication.

#### 3. Data Visualization
- **Heatmaps**: Matrices can be visualized as heatmaps, where each cell's color intensity represents the value of the corresponding matrix element.
- **Scatter Plots and Histograms**: Specific columns (properties) can be selected and visualized to understand distributions and relationships.

### Mathematical Foundations

Matrices provide a compact and powerful way to handle and manipulate data. Here are a few mathematical operations that are commonly used:

#### 1. Matrix Multiplication
Matrix multiplication allows combining datasets or transforming data through operations like rotation and scaling. For instance, multiplying a data matrix by a transformation matrix can change the dataset's scale or orientation.

#### 2. Eigenvalues and Eigenvectors
These are used in principal component analysis (PCA) to reduce the dimensionality of data while preserving its variance, which is crucial for visualizing high-dimensional data and improving computational efficiency in machine learning algorithms.

#### 3. Singular Value Decomposition (SVD)
SVD is used in various applications, including noise reduction, data compression, and collaborative filtering (e.g., recommendation systems).

### Conclusion

Understanding data as matrices with groups (rows) and properties (columns) provides a structured and mathematical approach to data representation and manipulation. This perspective is foundational in fields like data science, machine learning, and statistics, enabling efficient data analysis, transformation, and visualization.
