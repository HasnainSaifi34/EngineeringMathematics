```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("./data/countries.csv")
df = df.dropna()

RankMatrix = np.array(
    [
        df["2015 Rank"],
        df["2014 Rank"],
        df["2013 Rank"]
    ]
)

ScoreMatrix = np.array(
    [
        df["2015 Score"],
        df["2014 Score"],
        df["2013 Score"]
    ]
)


CountryName = df["Country Name"]

sample = 6

# Prepare data for sampled countries
Sampled_Country_Name = CountryName[:sample]  # Select first 'sample' countries
Sampled_Score_Matrix = ScoreMatrix[:, :sample]  # Select corresponding scores for all years

# Define positions for bars on the x-axis (centered)
country_positions = np.arange(len(Sampled_Country_Name))

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as desired

# Create bars for each year
plt.bar(country_positions - 0.2, Sampled_Score_Matrix[0], width=0.4, label="2015")
plt.bar(country_positions, Sampled_Score_Matrix[1], width=0.4, label="2014")
plt.bar(country_positions + 0.2, Sampled_Score_Matrix[2], width=0.4, label="2013")

# Set labels and title
plt.xticks(country_positions, Sampled_Country_Name, rotation=45)  # Rotate x-axis labels for readability
plt.xlabel("Country")
plt.ylabel("Happiness Score")
plt.title("Happiness Score of Sampled Countries (2013-2015)")

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()

```
