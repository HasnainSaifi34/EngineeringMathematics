```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

----
# Implementing Gradient Descent on a 1D function 

-----

```python
s_x = sym.var('x')
s_y =  3*s_x**2-3*s_x+4
s_ydy = 6*s_x-3;
xDomain = [-2,2]
steps = 5000
xx = np.linspace(xDomain[0],xDomain[1],steps)
tolerance = 1e-8
fx = sym.lambdify(s_x,s_y)
fxdx = sym.lambdify(s_x,s_ydy)

# applying gradient descent algorithm to find the minimum
# fx(0.5) =3.25 is the minimum at x = 0.5
e = 0.0001
#x = 5*np.random.randn()
x=0
while(True):
  if abs(fxdx(x)) < tolerance:  # Check if derivative is close to zero
    print(f"the minima is reached at x = {x}")
    break;
  x = x  +  (- e*fxdx(x))   # if -C(W) --gradient is -ve which means x should "decrease"  by a magnitude ofx = x + -(e*C(W))
                            # if -C(W) --gradient is +ve which means x should "increase"  by a magnitude ofx = x + -(e*C(W))
                            

plt.plot(xx,fx(xx),label="f(x)")
plt.plot(xx,fxdx(xx),label = "$ (d/dx)f(X)  $")
plt.plot(fx(0.5),"o",label="minima = 3.25")
plt.legend()
plt.show()


```
