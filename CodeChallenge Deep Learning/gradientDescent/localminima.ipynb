## How Gradient Descent can fail if started at a wrong value

-----------

```python
import numpy as np; 

import matplotlib.pyplot as plt;
```

```python
a=2
xx = np.linspace(-1*a,+1*a,5000)

def fxdx(x,fx):
    h = 1e-10;
    value =( (fx(x+h)-fx(x))/h )
    return value;

def fx(x):
    value = np.cos(2*np.pi*x) +x*x
    return value



plt.plot(xx,fx(xx))
plt.plot(xx,fxdx(xx,fx))



```
