```python
import numpy as np;
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
x_y_sample = []

xx = np.linspace(-1,+1,40).reshape(-1,+1)

for i in range(len(xx)):
    x_y_sample.append([xx[i],np.sin(2*xx[i])])



x = [item[0] for item in x_y_sample]
y = [item[1] for item in x_y_sample]


plt.scatter(x,y)
```

```python
# Generate dataset
x_values = np.linspace(-10, 10, 100).reshape(-1, 1)
y_values = np.sin(2 * x_values)

# Check shapes of x_values and y_values
print(f"x_values shape: {x_values.shape}")
print(f"y_values shape: {y_values.shape}")

# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),  # 1 hidden layer with 10 neurons
    tf.keras.layers.Dense(1, activation='relu')  # Output layer with linear activation for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(x_values, y_values, epochs=4000, verbose=0)

# Make predictions
predictions = model.predict(x_values)

# Ensure predictions have the correct shape
predictions = predictions.reshape(-1, 1)

# Plot the results
plt.plot(x_values, y_values, label='True Function')
plt.plot(x_values, predictions, label='Neural Network Approximation')
plt.legend()
plt.show()
```
