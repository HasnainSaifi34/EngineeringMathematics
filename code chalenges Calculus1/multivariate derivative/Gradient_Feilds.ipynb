```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define the function and its gradient
def f(x, y):
    return x**2 + y**2

# Gradient of f with respect to x
def Grad(x,y):
  return [2*x,2*y]

step=10
# Create a grid of points
x = np.linspace(-10, 10, step)
y = np.linspace(-10, 10, step)
X, Y = np.meshgrid(x, y)

# Calculate the gradient vector at each point
U = Grad (X,Y)[0]
V = Grad (X,Y)[1]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the gradient vectors at each point
ax.quiver(X, Y, U, V, color='r', width=0.002)

# Set limits and labels for axes
ax.set_xlim(-11, 11)
ax.set_ylim(-11, 11)
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Show the plot
plt.show()
```

```python
step=6000
# Create a grid of points
x = np.linspace(-10, 10, step)
y = np.linspace(-10, 10, step)
X, Y = np.meshgrid(x, y)

# plot the 3d funtion
Z = f(X,Y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
```

```python
plt.contourf(X, Y, f(X, Y))
plt.colorbar()
plt.show()
```

```python
step=100
# Create a grid of points
x = np.linspace(-10, 10, step)
y = np.linspace(-10, 10, step)
X, Y = np.meshgrid(x, y)

# Calculate the gradient vector at each point (using the same 'step')
U = Grad(X,Y)[0]
V = Grad(X,Y)[1]


fig , ax = plt.subplots(1,2,figsize=(12, 6))

ax[0].contourf(X, Y, f(X, Y))
ax[0].set_title('Contour Plot')

ax[1].quiver(X, Y, U, V, color='r', width=0.0060) # Now U and V match X and Y
ax[1].set_title('Gradient Field')


plt.show()
# plot the function
```
