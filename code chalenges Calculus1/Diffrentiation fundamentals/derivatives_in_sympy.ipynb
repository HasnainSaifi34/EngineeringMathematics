## Implementing derivatives in sympy 
$ f\left(x\right)\ =\ \frac{5}{4}x+\frac{9}{4} $

----



```python
import sympy as sym;
from sympy.abc import x
```

```python
fx = (5/4)*x + 9/4

fx
```

```python
sym.plot(fx,(x,-1,+6),axis_center=[0,0])
```

```python
dfx = sym.diff(fx,x)

print(f"the derivative of the function is {dfx}")
```

## Derivative of $ f(x) = x^2 $ in sympy

-----

```python
fx = x**2;

fx
```

```python
sym.plot(fx,(x,-1,6),axis_center=[0,0])
```

```python
df_dx = sym.diff(fx,x)

df_dx
```

```python
sym.plot(df_dx , (x,-1,6))
```
