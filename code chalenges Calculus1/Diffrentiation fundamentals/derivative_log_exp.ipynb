```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, +7, 6000)


x1 = 4
h = 1e-8  # Small value to approximate the secant line
x2 = x1 + h

y1 = np.log(x1)
y2 = np.log(x2)

# Slope of the secant line
m = (y2 - y1) / (x2 - x1)

# Equation of the secant line
secant_line = y1 + m * (xx - x1)
fig, ax = plt.subplots(1, 2, figsize=(24, 8))

# Plot the sine function and the secant line
ax[0].plot(xx, np.log(xx), label='$ ln(x) $', color='blue')
ax[0].plot(xx, secant_line, label='tangent Line', color='red',linewidth=2)
ax[0].scatter([x1, x2], [y1, y2], color='green', zorder=5)  
ax[0].set_title(f'$ ln(x) $ secant at x={x1} and y={y1}')
ax[0].axhline(0,color="g",linestyle="--")
ax[0].axvline(0, color="g",linestyle="--")
ax[0].set_ylim(-5,+5)
# Plot the cosine function
ax[1].plot(xx, (1/xx), label="f'(x) = 1/x", color='black')
ax[1].set_title(f"f'(x) = 1/x value at x={x1} and y = {1/x1}")
ax[1].axvline(x1, color="red", linestyle="-")# Vertical line at x1
ax[1].scatter(x1,(1/x1))
ax[1].set_ylim(-4,+400)

# plot the origin
ax[1].axhline(0,color="g",linestyle="--")
ax[1].axvline(0, color="g",linestyle="--")

plt.legend()
plt.show()
```

# Derivative $ ln(x) $ and $log_{a}(x) $

## $ f(x) = ln(x) $ then $ f'(x) = \frac{1}{x} $ for $ x<0 $

## $ f(x) = log_{a}(x) $ then $ f'(x) = \frac{1}{xln(a)} $


---

```python
xx = np.linspace(0,5,6000)


log2 = lambda x : np.log2(x)

log10 = lambda x : np.log10(x)

loge = lambda x : np.log(x)

def dfdx(x,base):
    return ( 1 / x*np.log(base) )

fig , ax = plt.subplots(1,2,figsize=(24,8))
ax[0].plot(xx, log2(xx), label=r"$\log_2(x)$", color="blue", linewidth=2)
ax[0].plot(xx, log10(xx), label=r"$\log_{10}(x)$", color="green", linewidth=2)
ax[0].plot(xx, loge(xx), label=r"$\ln(x)$", color="red", linewidth=2)
ax[0].set_ylim(-2, 3)
ax[0].set_title("Logarithmic Functions", fontsize=18)
ax[0].set_xlabel("x", fontsize=16)
ax[0].set_ylabel("log(x)", fontsize=16)
ax[0].grid(True)
ax[0].legend(fontsize=14, loc='upper left')

ax[1].plot(xx, dfdx(xx, 2), label=r"$f'(x)$ of $\log_2(x)$", color="blue", linestyle="--", linewidth=2)
ax[1].plot(xx, dfdx(xx, 10), label=r"$f'(x)$ of $\log_{10}(x)$", color="green", linestyle="--", linewidth=2)
ax[1].plot(xx, dfdx(xx, np.exp(1)), label=r"$f'(x)$ of $\ln(x)$", color="red", linestyle="--", linewidth=2)
ax[1].set_ylim(0, 5)
ax[1].set_title("Derivatives of Logarithmic Functions", fontsize=18)
ax[1].set_xlabel("x", fontsize=16)
ax[1].set_ylabel(r"$f'(x)$", fontsize=16)
ax[1].grid(True)
ax[1].legend(fontsize=14, loc='upper right')

plt.suptitle("Logarithmic Functions and Their Derivatives", fontsize=22)
plt.legend()
plt.show()

```

## investigating derivative of $ \frac{d}{dx} e^x $


---

```python

xx = np.linspace(-10, 10, 6000)

h = 1e-8

x1 = 2.5
x2 = x1 + h
y1 = np.exp(x1)
y2 = np.exp(x2)

m = (y2 - y1) / (x2 - x1)

secant_line = y1 + m * (xx - x1)

plt.figure(figsize=(10, 6))

# Plot the exponential function
plt.plot(xx, np.exp(xx), label=r"$\exp(x)$", color="blue", linewidth=2)

# Plot the secant line
plt.plot(xx, secant_line, label="Secant Line", color="red", linestyle="--", linewidth=2)

# Highlight the point of tangency
plt.scatter([x1], [y1], color="green", s=100, zorder=5, label=f"Point ({x1:.2f}, {y1:.2f})")

# Adding labels and titles
plt.ylim(-20, 20)
plt.title(f"slope of tangent = {m} exp({x1})={y1} ", fontsize=12)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

# Grid and legend
plt.grid(True)
plt.legend(fontsize=12)

plt.show()
```

# through this experiment we get to know that the slope of tangent line at x is same as the function value at x


# which means $ \frac{d}{dx} e^x = e^x $

# $\frac{d(e^{cx})}{dx} = ce^{cx} $


# lets go through the derivatives of exponential functions in detail

- ## lets see the case of $ f(x) = a^x $

### $ f'(x) = lim_{h->0} [ \frac{(a^{x+h} - a^{x} )}{h} ] $

### $ f'(x) = lim_{h->0} [ \frac{(a^{x}a^{h} - a^{x} )}{h} ] $
### $ f'(x) = lim_{h->0} a^{x}[ \frac{(a^{h} - 1 )}{h} ] $

### $ f'(x) = a^{x} lim_{h->0} [ \frac{(a^{h} - 1 )}{h} ] $

### let $ \text{k} $ = $lim_{h->0} [ \frac{(a^{h} - 1 )}{h} ]$ 

## $ f'(x) = f(x)k $ or $ \frac{d(a^x)}{dx} = a^x k $ 
----

## Exponential functions have an amazing property where their derivatives is proportional to the original function

## $ \frac{d(a^x)}{dx} = a^x k $  and $ k = ln(a) $


## soo what $ k = 1$ such that $ \frac{d(a^x)}{dx} = a^x $ 

## then $ a = e$ and $ ln(e) = 1 $

```python
xx = np.linspace(-6, +6, 6000)

# Constants
a1, a2, a3 = (3.8, 3, np.exp(1))

# Power function
power_function = lambda a, power: a ** power

# Create subplots
fig, ax = plt.subplots(1, 3, figsize=(24, 8))

# First subplot
ax[0].plot(xx, power_function(a1, xx), label=f"${a1}^x$", color="red", linewidth=2.5)
ax[0].plot(xx, power_function(a1, xx) * np.log(a1), label=(r"$\frac{d(3.8^x)}{dx} = 3.8^xk $"  + f" ,  k = {np.log(a1)}"), linestyle="--", linewidth=2)
ax[0].set_title(f"Function and Derivative of {a1}^x", fontsize=16)
ax[0].set_xlabel("x", fontsize=14)
ax[0].set_ylabel("y", fontsize=14)
ax[0].legend(fontsize=16)
ax[0].grid(True)

# Second subplot
ax[1].plot(xx, power_function(a2, xx), label=f"${a2}^x$", color="green", linewidth=2.5)
ax[1].plot(xx, power_function(a2, xx) * np.log(a2), label=(r"$\frac{d(3^x)}{dx} = 3^xk $"  + f" ,  k = {np.log(a2)}"), color="lime", linestyle="--", linewidth=2)
ax[1].set_title(f"Function and Derivative of {a2}^x", fontsize=16)
ax[1].set_xlabel("x", fontsize=16)
ax[1].legend(fontsize=14)
ax[1].grid(True)

# Third subplot
ax[2].plot(xx, power_function(a3, xx), label=f"$y = e^x$", color="blue", linewidth=2.5)
ax[2].plot(xx, power_function(a3, xx) * np.log(a3), label=(r"$\frac{d(e^x)}{dx} = e^xk $"  + f" ,  k = {np.log(a3)}"), color="cyan", linestyle="--", linewidth=2)
ax[2].set_title(f"Function and Derivative of $e^x$", fontsize=16)
ax[2].set_xlabel("x", fontsize=16)
ax[2].legend(fontsize=14)
ax[2].grid(True)

plt.tight_layout()
plt.show()
```
