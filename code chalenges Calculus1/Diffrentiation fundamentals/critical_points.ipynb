```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fx = lambda x : x**2 + x**3
# x lim -1.5 , 1 
# y lim -1 , 2
fig , ax = plt.subplots(2,1,figsize=(12,10))
h = 1e-4
xx = np.arange(-1.5,1,h)
yy = fx(xx)

dy = np.diff(yy)
dx = xx[1] - xx[0]
dy_dx = dy/dx

ax[0].plot(xx,fx(xx))

ax[0].set_ylim(-1,2)

ax[1].plot(xx[:-1],dy_dx)
ax[1].set_ylim(-1,2)
ax[1].axhline(0,color="red")


plt.tight_layout()
plt.show()
```

```python
from scipy.signal import find_peaks

dflocalmin = find_peaks(-np.abs(dy_dx))[0] 

```

```python
fig , ax = plt.subplots(2,1,figsize=(12,10))
ax[0].plot(xx,fx(xx))
ax[0].scatter(xx[dflocalmin],fx(dflocalmin),color="red" ) 

ax[0].set_ylim(-1,2)

ax[1].plot(xx[:-1],dy_dx)
ax[1].set_ylim(-1,2)
ax[1].axhline(0,color="red")


plt.tight_layout()
plt.show()
```
