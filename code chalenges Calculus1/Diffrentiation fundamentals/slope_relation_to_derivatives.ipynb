```python
import numpy as np;
import matplotlib.pyplot as plt
```

$ m=\frac{\Delta y}{\Delta x} $ is the slope of a linear function </br>

$ f\left(x\right)\ =\ mx\ +\ b $

this slope describes the change in output based on change in input

```python
point1 = [-1 , 1]

point2 = [3 , 6]

m = ( point2[1] - point1[1] ) / ( point2[0] - point1[0] )

b = 2.25
def f(x):
    return m*x + b

xx = np.linspace(-1,3,6000)

plt.plot(xx,f(xx))
plt.plot(point1[0],point1[1],"ro",label="p1")
plt.plot(point2[0],point2[1],'bs',label ="p2")
plt.xlim(-2,7)
plt.ylim(-2,7)
plt.axvline(x=0)
plt.axhline(y=0)
plt.title(f"the slope of line is m = {m}")
plt.legend()
plt.show()

```

## Sample $ f(x) = x^2 $  at N = 5

and calculate the slope in each coresponding point

and plot the slope series


-----

```python
N = 5;
start = 1.5
x = []
y = []
for i in range(N):
   x.append(i+start) 
   y.append(f(x[i]))

def f(x):
    return x**2;


plt.plot(x,y)
plt.title("sample of f(x) = x^2 at n=5")
for i in range(len(x)):
    plt.plot(x[i],y[i],"bo")

```

```python
slopes = []

for i in range(len(x)-1):
    value = (y[i+1] - y[i])/(x[i+1] - x[i])
    slopes.append(value)
    
 

for i in range(len(slopes)):
    plt.hlines(y=slopes[i], xmin=x[i], xmax=x[i + 1], colors='green')

# Customize plot limits
plt.xlim(1.5, 9)
plt.ylim(0,25)
plt.xlabel('x')
plt.ylabel('Slopes')
plt.title('Horizontal Lines Representing Slopes')

# Display the plot
plt.show()
```

## Sample $ f(x) = x^2 $  at N = 50

and calculate the slope in each coresponding point

and plot the slope series


-----

```python
N = 50;
start = 1.5
x = []
y = []
for i in range(N):
   x.append(i+start) 
   y.append(f(x[i]))

def f(x):
    return x**2;


plt.plot(x,y)
plt.title("sample of f(x) = x^2 at n=50")
for i in range(len(x)):
    plt.plot(x[i],y[i],"bo")
```

```python
slopes = []

for i in range(len(x)-1):
    value = (y[i+1] - y[i])/(x[i+1] - x[i])
    slopes.append(value)
    
 



fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot horizontal lines in the first subplot
for i in range(len(slopes)):
    ax[0].hlines(y=slopes[i], xmin=x[i], xmax=x[i + 1], colors='green')

# Customize plot limits and labels for the first subplot
ax[0].set_xlim(0, 100)
ax[0].set_ylim(-1, max(slopes) + 5)
ax[0].set_xlabel('x')
ax[0].set_ylabel('Slopes')
ax[0].set_title('Horizontal Lines Representing Slopes')

# Plot a line in the second subplot
xx = np.linspace(0, 100, 6000)
ax[1].plot(xx, 2 * xx, color='blue')

# Customize plot limits and labels for the second subplot
ax[1].set_xlim(0, 100)
ax[1].set_ylim(0, 200)
ax[1].set_xlabel('xx')
ax[1].set_ylabel('2*xx')
ax[1].set_title("f'(x) the derivative of the function")

# Display the plot
plt.tight_layout()
plt.show()
```
