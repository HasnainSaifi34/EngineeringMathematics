<h2> <b>Plot 3 Graphs f1 , f2 , f3 </b></h2>
<h3> <i> f1 = sin(cos(x)) fs = cos(sin(x)) f3 = cos(x) </i></h3></br>

<h3>Dx:(-pi,2pi)</h3>

```python
import numpy as np;
import sympy as sym;
import matplotlib.pyplot as plt;
```

```python
xDomain = [-np.pi,2*np.pi]

s_x = sym.var('theta')

s_y1 = sym.sin(sym.cos(s_x))
s_y2 = sym.cos(sym.sin(s_x))
s_y3 = sym.cos(s_x)

fx1 = sym.lambdify(s_x,s_y1);
fx2 = sym.lambdify(s_x,s_y2);
fx3 = sym.lambdify(s_x,s_y3);
x = np.linspace(xDomain[0],xDomain[1],400);
plt.plot(x,fx1(x),label="sin(cos(x))")
plt.plot(x,fx2(x),label="cos(sin(x))")
plt.plot(x,fx3(x),label="cos(x)")
plt.legend();
plt.grid();
plt.xlim(xDomain[0],xDomain[1])
plt.show()


```
