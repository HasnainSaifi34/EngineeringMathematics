# Implementation of Piecwise Functions in Numpy 


----


```python
import numpy as  np;
import matplotlib.pyplot as plt;
import sympy as sym;
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
xDomain = [-3,5]
steps = 100
xx = np.linspace(xDomain[0],xDomain[1],steps)


part1 = 0         * (xx<0)
part2 = -2*xx     * ((xx <= 0) &  (xx <3))
part3 = 0.1*xx**3 * (xx>=3)


y = part1+part2+part3;
plt.plot(xx,y,label="piecewise.ipynb")
plt.grid(True)
plt.legend()
plt.show()
```

----

<h2> Plot these functions seperately </h2>

----

```python
# plot the function pieces
plt.plot(xx[xx<0]             , np.zeros(np.sum(xx<0))  ,linewidth=2,label='Piece 1')
plt.plot(xx[(xx>=0) & (xx<3)] , -2*xx[(xx>=0) & (xx<3)] ,linewidth=2,label='Piece 2')
plt.plot(xx[xx>=3]            , xx[xx>3]**3/10          ,linewidth=2,label='Piece 3')

# make the plot look a bit nicer
plt.legend()
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.title('A piecewise function')
plt.show()
```

----
<h2>Plot these function using sympy </h2>

----

```python
from sympy.abc import x

# list function pieces
piece1 = 0
piece2 = -2*x
piece3 = x**3/10

# put them together with conditions
fx = sym.Piecewise( (piece1,x<0),(piece2,(x>=0) & (x<3)),(piece3,x>=3) )

# plot it
sym.plot(fx,(x,xx[0],xx[-1]),title='Piecewise function',
         xlabel='x',ylabel='y');
```

----

<h2> RELU Function in sympy </h2>

----


```python
piece1 = 0;
piece2 = x;
xDomain = [-3,3]
s_y = sym.Piecewise((0, x < 0), (x, True))
xx = np.linspace(xDomain[0],xDomain[1],steps)
fx = sym.lambdify(x,s_y);

plt.plot(xx,fx(xx))
plt.grid()
plt.xlim(xDomain[0],xDomain[1])
fx(10)
```

----
<h2> Leaky RELU in sympy and numpy </h2>

----

```python
piece1 = 0.01*x;
piece2 = x;
xDomain = [-3,3]
s_y = sym.Piecewise((piece1, x < 0), (x, True))
xx = np.linspace(xDomain[0],xDomain[1],steps)
fx = sym.lambdify(x,s_y);
plt.plot(xx,fx(xx))
plt.xlim(xDomain[0],xDomain[1])
plt.grid()
plt.show()
fx(-3)
```
