# This includes plotting problems using numpy and smypy

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xDomain = [-2,2]
#Create a Domain
resolution = .0001
#x = np.arange(xDomain[0],xDomain[1],resolution)
# [-2 ,2)

#option 1
x = np.arange(xDomain[0],xDomain[1]+resolution,resolution)
#[-2,2]

#option2
x = np.linspace(xDomain[0],xDomain[1])

# creating out function y = x^2+3x^3-x^4

y = (x**2)+(3*x**3)-(x**4)

y
```

<h2><b>Plotting the Function y = x<sup>2</sup> + 3x<sup>3</sup> - x<sup>4</sup></h2>

```python
plt.plot(x,y,label=f"$y = x^2+3x^3-x^4$")
plt.xlim(xDomain[0],xDomain[1])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

----
<h2>  Implementing the function <b>f(x) = -x<sup>4</sup>+3x<sup>3</sup>+x<sup>2</sup>  </b>  in <i>Scmpy</i></h2> </br>


----
<h3> Using Sympy we can create a symbolic variable rather than a python variable </h3>

----

```python
# Creating a Symbolic Variable
# beta means x

s_beta = sym.var('beta')

#defiing the function

s_y = (s_beta**2)+(3*(s_beta**3))-(s_beta**4)
s_y
```

```python
sym.plot(s_y,(s_beta,xDomain[0],xDomain[1]),xlabel="x",ylabel=f"f({s_beta})",title=f"$f(\\beta) ={sym.latex(s_y)}$" )
```

----
# Converting sympy to numpy 

<h1><b> Why Convert ? </b> : Because of Sympy Plot Engine Limitation </h1>

<h3> this is necessary to work with symbolic functions and to plot them </h3>

----



```python
fx = sym.lambdify(s_beta,s_y)
fx(2)
# this will return the function value at f(2)

beta = sym.latex(s_beta)
# this will return the latex version of beta
plt.plot(x,fx(x),label=f"$y = -{beta}^4+3{beta}^3+{beta}^2$")
plt.xlim(xDomain[0],xDomain[1])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# x,y are numpy variables defined above
```
