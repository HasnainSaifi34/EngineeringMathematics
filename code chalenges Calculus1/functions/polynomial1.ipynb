----

<h3> Generate a 3rd Order Polynomial  with random coeficients</h3>

----

```python
# a n =3 no of coeficient = n+1
class PolyNomials:
    def __init__(self,s_x,order):
        self.x = s_x;
        self.n = order;
    
    def GenerateCoefiencts(self):
         coeficients = np.random.randn(self.n+1)
         return coeficients
    def GeneratePolynomial(self):
       coef = self.GenerateCoefiencts()
       y = 0;
       for i in range(len(coef)):
        y += coef[i]*self.x**(i)
       return y
    
s_x = sym.symbols('x')
order = 3;
obj = PolyNomials(s_x,order)

s_y = obj.GeneratePolynomial()
fx = sym.lambdify(s_x, s_y, modules=['numpy'])
fx(2)

```

```python
# Define the domain for x
xDomain = [-100, 100]
x = np.linspace(xDomain[0], xDomain[1], 400)

# Plot the function
LatexEquation = sym.latex(s_y)
plt.plot(x, fx(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Plot of a Random Polynomial of Order :f(x) =${LatexEquation}$ {order}')
plt.grid(True)
plt.show()
```
