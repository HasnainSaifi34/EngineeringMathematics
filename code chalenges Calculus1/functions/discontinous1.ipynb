<h3> Jump Discontinuity in Numpy </h3>
<h2>
f(x) = {</br>
 "   "sin(x*pi) if(x<0) </br>
 "   "1.5 if(x==0) </br>
 "   "-(x-2)<sup>2</sup> if(x>0)
}
</h2>


----

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xDomain= [-1,2]
steps = 0.1


xx = np.arange(xDomain[0],xDomain[1],steps)

# list of the function pieces 
y = [0]*3

y[0]=np.sin(xx*np.pi)

y[1] = 1.5*np.ones(len(xx))

y[2] = -(xx-2)**2 


domain_list = [0]*3

domain_list[0] = xx<0;
domain_list[1]  = np.abs(xx)<steps/10 ; # this is going to tell us the one point that is clossest to zero
domain_list[2] = xx>0;

marker ="-o-"
for i in range(len(y)):
    plt.plot(xx[domain_list[i]],y[i][domain_list[i]],marker[i])
```

# Implementing the same function using sympy 


------


```python
from sympy.abc import x;
piece1 = sym.sin(sym.pi*x)
piece2 = 1.5;
piece3 = -(x-2)**2

fx = sym.Piecewise(
    (piece1,x<0),
    (piece2,sym.Eq(x,0)),
    (piece3,x>0)
)
fx.subs(x,0)
sym.plot(x,fx,(x,xx[0],xx[-1])) # (x,xx[0],xx[-1]) = domain values of the funtion
fx
```
