# Excercise No.3
just working with composite functions in sympy and not in numpy

$ f\left(x\right)\ =\ \sin\left(x\right) $


$ g\left(x\right)\ =\ \ln\left(x\right) $

$ h\left(x\right)\ =\ 2x^{2}+5 $

Find: $ f\left(g\left(h\left(x\right)\right)\right) $

-----

```python
import sympy as sym;
import matplotlib.pyplot as plt
```

```python
from sympy.abc import x;

fx = sym.sin(x);

gx = sym.ln(x);

hx = 2*(x**2)+5;

# fx.subs(x,hx) -> sin(2x^2+5)

composite =fx.subs(x,gx.subs(x,hx))
xDomain=[-100,+100]
sym.plot(composite,(x,xDomain[0],xDomain[1]),title=f"$y = {sym.latex(composite)} $")

```
