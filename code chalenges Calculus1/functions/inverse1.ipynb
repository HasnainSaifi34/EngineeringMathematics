-----
$ f\left(x\right)\ =\ \ln\left(2x\right) $

$ g\left(x\right)\ =\ \frac{e^{x}}{2} $   <b> g(x) is f^-1(x) </b>

</br>

Find  $ f\left(g\left(x\right)\right) $ and $ g\left(f\left(x\right)\right) $




-----

```python
import numpy as np;
import matplotlib.pyplot as plt;


```

```python
def f(x):
    value = np.log(2*x);
    return value;

def g(x):
    value = (np.exp(x))/2
    return value;

def gof(x):
    value = g(f(x))
    return value;

def fog(x):
    value =f(g(x))
    return value;

xDomain = [1,5]
yDomain = [-4,10]

xx = np.linspace(xDomain[0],xDomain[1],30)

plt.plot(xx,f(xx),label="f(x)")
plt.plot(xx,g(xx),label="g(x)")
plt.plot(xx,fog(xx),label="f(g(x))")
plt.plot(xx,gof(xx),"o",label="g(f(X))")
plt.ylim(yDomain[0],yDomain[1])
plt.legend()
plt.show()
```
