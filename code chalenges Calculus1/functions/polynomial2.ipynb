## Maclaurin Series

The Maclaurin series, named after Scottish mathematician Colin Maclaurin, is a representation of a function as an infinite sum of its derivatives evaluated at zero. It's a powerful tool in calculus for approximating functions and understanding their behavior around a specific point (usually x = 0).

### Formula

The general formula for the Maclaurin Series of a function f(x) is:

$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}}(0) \frac{x^n}{n!}$$

where:

* f(x) is the function we want to represent as a series.
* f^(n)(0) is the nth derivative of f(x) evaluated at x = 0 (nth order derivative).
* x is the variable.
* n! is the factorial of n (n * (n-1) * ... * 1).
* The summation symbol (Î£) indicates summation over n from 0 to infinity.

### Explanation

The Maclaurin series essentially breaks down the function f(x) into a series of simpler terms. Each term involves the nth derivative of f(x) evaluated at zero, multiplied by the variable x raised to the power of n, all divided by the factorial of n. As we add more terms (higher values of n), the accuracy of the approximation for f(x) increases.

Here are some key points to remember:

* The Maclaurin series converges (approaches the actual function) for some functions only within a specific interval around x = 0.
* Common functions like e^x, sin(x), cos(x), and ln(1+x) have well-known Maclaurin series expansions that can be used for calculations.

By understanding the Maclaurin series formula and its applications, you can approximate functions, analyze their behavior near specific points, and gain deeper insights into their properties.


```python
# a = number of terms to sum up
#we need to sum up n=1 to n=a 
class MaclaurinSeries:
    def __init__(self,a):
        self.a = a;
        # it tracks indivisual terms
        self.terms = []

    def Terms(self,n):
      # this function will return term for a given n value
      s_x = sym.var('x')
      v = 2*n-1
      s_y = (-1)**(n+1)* ((s_x**v)/np.math.factorial(v))
      return s_y
    
    def Sin(self):
      s_y = 0;
     
      for i in range(1,self.a+1):
         self.terms.append(self.Terms(i))
         s_y += self.Terms(i)
      return s_y  



```

```python
# this will plot the macluarin series

s_x = sym.var('x')
# a = no of terms
a = 100
obj = MaclaurinSeries(a)
# s_y approximately =  sin(x)
xDomain = [-50,50]
s_y = obj.Sin()
terms = obj.terms
x = np.linspace(xDomain[0],xDomain[1],400)
fx = sym.lambdify(s_x,s_y,modules=["numpy"])
LatexEquation = sym.latex(s_y)

# this will plot all the terms in the series
def PlotTerms():
  for i in range(a):
    function = sym.lambdify(s_x,terms[i])
    plt.plot(x,function(x),'--', label=f'Term {i+1}')

# PlotTerms()    
# this will plot the series
plt.plot(x, fx(x),color="red")    
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Plot of a Random Polynomial of Order :f(x) =$Macluarin Series$ ')
plt.xlim(xDomain[0],xDomain[1])
plt.grid(True)
plt.show()

```
