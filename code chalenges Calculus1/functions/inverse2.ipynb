# Sympy Excercise <b> Finding the inverse in sympy </b>


$f\left(x\right)\ =\ 2x+3 $

$ g\left(x\right)\ =\ 2x\ +\sin\left(x\right)$

Demonstrate $f^{-1}\left(f\left(4\right)\right) $ = $ g^{-1}\left(g\left(4\right)\right)$ = 4




----

```python
import sympy as sym
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
x = sym.var('x')
y = sym.var('y')
fx = 2*x+3;
gx = 2*x + sym.sin(x)



equation1 = sym.Eq(y,fx)
equation2 = sym.Eq(y,gx)


inverse1 =sym.solve(equation1,x)[0] # this function will solve the equation for x
inverse1 =  inverse1.subs(y,x)
a = 4;

# f^-1(f(4)) -> implement this

value = inverse1.subs(x,fx.subs(x,a))

print("the value at f^-1(f(4)) :",value)


# inverse2 = sym.solve(equation2,x)[0] this function is not invertible soo sympy will raise error







xx = np.linspace(-10,+10,5000)
fx = sym.lambdify(x,fx)
inv_fx = sym.lambdify(x,inverse1)
plt.plot(xx,fx(xx))
plt.plot(xx,inv_fx(xx))





```
