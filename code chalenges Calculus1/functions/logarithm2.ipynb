f(x) = e<sup>x</sup> - ln(x) - e </br>
 e<sup>x</sup> = sym.exp()</br>

```python
import numpy as np;
import sympy as sym;
import matplotlib.pyplot as plt;
```

```python
beta = sym.var('beta')
s_y = sym.exp(beta)- sym.log(beta) - np.exp(1)
fx = sym.lambdify(beta,s_y)
xDomain = [0,10]
x = np.linspace(xDomain[0],xDomain[1])
latex = sym.latex(s_y)
print("sympy plot")
sym.plot(s_y)
print("Numpy Plot")
plt.plot(x,fx(x),label=f"$f(\\beta) = {latex}$")
plt.xlim(xDomain[0],xDomain[1])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
s_x = sym.var('x');
s_y1 = sym.exp(s_x);
s_y2 = sym.log(s_x);
s_y3 = sym.log(sym.exp(s_x))
s_y4 = sym.exp(sym.log(s_x))
s_y = [s_y1,s_y2,s_y3,s_y4]
xDomain = [-4,4]
x = np.linspace(xDomain[0],xDomain[1],400);

for i in range(len(s_y)):
    fx = sym.lambdify(s_x,s_y[i]);
    plt.plot(x,fx(x))

    plt.axis('square')

plt.xlim([xDomain[0],xDomain[1]])
plt.ylim([xDomain[0],xDomain[1]])
plt.show()




```
