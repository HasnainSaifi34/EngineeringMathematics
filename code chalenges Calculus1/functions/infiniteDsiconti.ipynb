## Oscillating Discontinuity in f(x) = $f\left(x\right)\ =\ \sin\left(\frac{1}{\left(x-1\right)}\right)$

The function f(x) = sin(1/x - 1) exhibits an oscillating discontinuity, which means it has a break in its continuity due to a specific behavior around a certain point. Let's delve into the details:

**1. Discontinuity at x = 0:**

The term 1/x in the function creates an issue at x = 0. Division by zero is undefined, making the function itself undefined at that point.

**2. Oscillatory Behavior:**

As x approaches 0 from either the positive or negative side, the value of 1/x becomes very large (positive or negative) and fluctuates rapidly. This rapid change causes the argument of the sine function (1/x - 1) to oscillate between very large positive and negative values.

**3. No Defined Limit:**

Due to the constant oscillation of the sine function's argument as x approaches 0, the limit of f(x) as x approaches 0 doesn't exist. The function's values keep jumping between -1 and 1 infinitely as x gets closer and closer to 0, never settling on a single value.

**Key Points:**

* This discontinuity differs from a simple jump discontinuity where the function approaches different finite values from either side of the discontinuity point. 
* The rapid oscillation near x = 0 prevents the function from having a well-defined limit at that point.
* This behavior is characteristic of functions involving division by x or higher powers of x in the denominator.

**Visualizing the Discontinuity:**

Imagine graphing the function. There will be a gap at x = 0 because the function isn't defined there. As you zoom in closer to x = 0, the graph will appear to wiggle infinitely fast, never settling on a specific y-value. This rapid oscillation signifies the oscillating discontinuity.


```python
xDomain = [.99,1.01]
res = 5000;
xx = np.linspace(xDomain[0],xDomain[1],res)
a = 1

yy = a*np.sin(1/(xx-1))
plt.plot(xx,yy)
```
