# Excercise 1 : Estimate e lim n -> infinity (1+(1/n))^n
<h2> <b> lim <sub>k->0</sub> (1+k)<sup>(1/k)</sup></b></h2>
----

```python
import numpy as np;
import sympy as sym;
import matplotlib.pyplot as plt;
```

```python
class ApproxE:
    def __init__(self,terms):
        self.n = terms;
        self.TermsArray = []
        self.Approximation = []
        self.error = []
    def __TermsArray(self):
        for i in range(self.n):
            self.TermsArray.append(i+1)
    def __ApproximateFun(self):
       for i in range(len(self.TermsArray)):
           n = self.TermsArray[i]
           value = (1+(1/n))**n;
           self.Approximation.append(value)

        
    def __CalcError(self):
        e = np.exp(1);
        for i in range(len(self.Approximation)):
            value = e-self.Approximation[i]
            self.error.append(value)

    # i need to create a table with n | Ae | e-Ae

    def Table(self,display):
        self.__TermsArray() # it will generate numbers from 1 to n 
        self.__ApproximateFun()
        self.__CalcError()
        if(display):
         print(" n    |  e~    |   error")
         for i in range(len(self.TermsArray)):
            print(f" {self.TermsArray[i]}    |  {self.Approximation[i]}  |{self.error[i]}")

e = np.exp(1);

obj = ApproxE(5000)
obj.Table(True)

```

```python
s_x = sym.var('x');

s_y = max(obj.Approximation)**s_x;
print(max(obj.Approximation))
sym.plot(s_y)

x = np.linspace(-100,+100,400);
fx = sym.lambdify(s_x,s_y)
plt.plot(x,fx(x))
```

```python
s_n = sym.var('n');
# s_y = f(x) is an error function for all values of n on x axis  - actual e (approx - actual ) Error Function
x = np.linspace(0,+40,400);

apporxFx = ((1+(1/s_n))**s_n)
s_y = np.exp(1) - (apporxFx)

fx = sym.lambdify(s_n,s_y);
plt.plot(x,fx(x))
plt.ylabel("f(x) = e - approx(x)")
```
