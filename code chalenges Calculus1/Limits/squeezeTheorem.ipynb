The Squeeze Theorem, also known as the Sandwich Theorem or the Pinching Theorem, is a fundamental concept in calculus used to find the limit of a function. It states that if a function is "squeezed" between two other functions that have the same limit at a particular point, then the function itself must also approach that limit at that point.

### Formal Statement of the Squeeze Theorem

Let $ f(x) $, $ g(x) $, and $ h(x) $ be functions defined on an interval $ I $ (except possibly at a point $ a $). Suppose that for all $ x $ in $ I $ (where $ x \neq a $), we have:

$ g(x) \leq f(x) \leq h(x) $

If 

$ \lim_{{x \to a}} g(x) = \lim_{{x \to a}} h(x) = L $

then

$ \lim_{{x \to a}} f(x) = L $

### Visual Representation

Imagine the graph of $ g(x) $ and $ h(x) $ converging to the same value $ L $ as $ x $ approaches $ a $. The function $ f(x) $, being "squeezed" or "sandwiched" between $ g(x) $ and $ h(x) $, must also converge to the same value $ L $ as $ x $ approaches $ a $.

### Example

Consider the function $ f(x) = x^2 \sin\left(\frac{1}{x}\right) $ and we want to find the limit as $ x $ approaches 0.

1. We know that $\sin\left(\frac{1}{x}\right)$ oscillates between -1 and 1 for all $ x \neq 0 $.
2. Therefore, $ -1 \leq \sin\left(\frac{1}{x}\right) \leq 1 $.

Multiplying through by $ x^2 $, we get:

$ -x^2 \leq x^2 \sin\left(\frac{1}{x}\right) \leq x^2 $

As $ x $ approaches 0, both $ -x^2 $ and $ x^2 $ approach 0. By the Squeeze Theorem:

$ \lim_{{x \to 0}} x^2 \sin\left(\frac{1}{x}\right) = 0 $

Thus, the limit of $ f(x) = x^2 \sin\left(\frac{1}{x}\right) $ as $ x $ approaches 0 is 0.

### Applications

The Squeeze Theorem is particularly useful in cases where the function $ f(x) $ is difficult to evaluate directly at a certain point, but it can be bounded between two simpler functions whose limits are known and equal. This theorem is widely used in calculus, especially in the context of limits and continuity, and is a powerful tool for solving limit problems.

### Summary

The Squeeze Theorem helps us determine the limit of a function by "squeezing" it between two other functions with known limits. If the bounding functions converge to the same limit at a particular point, then the squeezed function must also converge to that limit.

```python
def g(x):
  return  .5 + (x-2)**2;

def f(x):
  return (( (x**2)-2*x )/( (x**2)-4 )  )

def h(x):
  return  .5 - (x-2)**2;


xx = np.linspace(-2,+4,6000)

plt.ylim(-10,+10)
plt.plot(xx,f(xx),label="f(x)")
plt.plot(xx,g(xx),label="g(x)")
plt.plot(xx,h(xx),label="h(x)")
plt.axvline(x=2, color='r', linestyle='--', label='x=2')

plt.legend()
plt.show()
```
