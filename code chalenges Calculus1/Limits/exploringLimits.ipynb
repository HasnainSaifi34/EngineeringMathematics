## Exploring Limits in Sympy and Numpy


-------

```python
import numpy as np 
import sympy as sym
import matplotlib.pyplot as plt
```


$
\lim_{{x \to 3}} f(x) = \lim_{{x \to 3}} \left( \cos(x \pi) + \left( \log(x) \right)^2 \right) 
 $

This notation represents the limit of the function $ f(x) = \cos(x \pi) + \left( \log(x) \right)^2 $ as $ x $ approaches 3.

```python
## Empirical Limits in Numpy
a = 6
xDomain =[0,6]
yDomain=[-1,6]

xx = np.linspace(xDomain[0],xDomain[1],6000)

def f(x):
    value = np.cos(x*np.pi)+np.log(x)**2
    return value

plt.plot(xx,f(xx))
plt.grid(True)
plt.ylim(yDomain[0],yDomain[1])
plt.show()
```

```python
x_y_postive=[ [2,f(2)]  ]
x_y_negetive=[  [4,f(4)] ]
xx = np.linspace(1,5,6000)
epoch = 0.05;

target = 3;

for i in range(0,20):
   if(x_y_postive[i][0]==target or x_y_negetive[i][0]==target):
      break;
   
   x_y_postive.append([x_y_postive[i][0]+epoch,f(x_y_postive[i][0]+epoch)])
   x_y_negetive.append([x_y_negetive[i][0]-epoch,f(x_y_negetive[i][0]-epoch)])

   
plt.ylim(-1,6)
plt.xlim(1,5)
plt.plot(xx,f(xx))  
# Plot positive points in blue ('bo') with label 'From the right'

# Plot negative points in red ('ro') with label 'From the left'
for i in range(len(x_y_negetive)):
    plt.plot(x_y_negetive[i][0], x_y_negetive[i][1], 'ro', label='From the right' if i == 0 else None)
    plt.plot(x_y_postive[i][0], x_y_postive[i][1], 'bo', label='From the left' if i == 0 else None)

plt.axvline(x=3, color='green', linestyle='--', label='Vertical line at x=3')

plt.grid(True)
plt.legend()
plt.show()


print(f"Function value at x=3 is {f(3)}")
print(f"Limit Approches  {x_y_negetive[-1][1]} from Right")
print(f"Limit Approches {x_y_postive[-1][1]} from left ")
```
