## Properties of Limits 


----

-----

In calculus, the concept of a limit is fundamental to defining derivatives, integrals, and continuity. Here are the key properties and rules of limits:

### Basic Properties of Limits
<h3>

1. **Limit of a Constant:**
   $
   \lim_{{x \to c}} k = k
   $
   For any constant $ k $, the limit as $ x $ approaches $ c $ is $ k $.

2. **Limit of the Identity Function:**
   $
   \lim_{{x \to c}} x = c
   $
   The limit of $ x $ as $ x $ approaches $ c $ is $ c $.

3. **Sum Rule:**
   $
   \lim_{{x \to c}} (f(x) + g(x)) = \lim_{{x \to c}} f(x) + \lim_{{x \to c}} g(x)
   $
   The limit of the sum of two functions is the sum of their limits.

4. **Difference Rule:**
   $
   \lim_{{x \to c}} (f(x) - g(x)) = \lim_{{x \to c}} f(x) - \lim_{{x \to c}} g(x)
   $
   The limit of the difference of two functions is the difference of their limits.

5. **Product Rule:**
   $
   \lim_{{x \to c}} (f(x) \cdot g(x)) = \lim_{{x \to c}} f(x) \cdot \lim_{{x \to c}} g(x)
   $
   The limit of the product of two functions is the product of their limits.

6. **Quotient Rule:**
   $
   \lim_{{x \to c}} \left( \frac{f(x)}{g(x)} \right) = \frac{\lim_{{x \to c}} f(x)}{\lim_{{x \to c}} g(x)}, \quad \text{provided} \quad \lim_{{x \to c}} g(x) \neq 0
   $
   The limit of the quotient of two functions is the quotient of their limits, provided the limit of the denominator is not zero.

7. **Power Rule:**
   $
   \lim_{{x \to c}} (f(x))^n = \left( \lim_{{x \to c}} f(x) \right)^n
   $
   The limit of a function raised to a power is the limit of the function raised to that power.

8. **Root Rule:**
   $
   \lim_{{x \to c}} \sqrt[n]{f(x)} = \sqrt[n]{\lim_{{x \to c}} f(x)}, \quad \text{provided} \quad \lim_{{x \to c}} f(x) \geq 0 \text{ for even } n
   $
   The limit of the $ n $-th root of a function is the $ n $-th root of the limit of the function, provided certain conditions are met for even roots.


</h3>



### Special Limits

1. **Squeeze Theorem:**
   If $ f(x) \leq g(x) \leq h(x) $ for all $ x $ near $ c $ (except possibly at $ c $) and
   $
   \lim_{{x \to c}} f(x) = \lim_{{x \to c}} h(x) = L,
   $
   then
   $
   \lim_{{x \to c}} g(x) = L.
   $

2. **One-Sided Limits:**
   - **Right-Hand Limit:**
     $
     \lim_{{x \to c^+}} f(x) = L
     $
     This is the limit of $ f(x) $ as $ x $ approaches $ c $ from the right.
   
   - **Left-Hand Limit:**
     $
     \lim_{{x \to c^-}} f(x) = L
     $
     This is the limit of $ f(x) $ as $ x $ approaches $ c $ from the left.

   For the two-sided limit to exist at $ c $, the right-hand limit and the left-hand limit must both exist and be equal:
   $
   \lim_{{x \to c}} f(x) = L \quad \text{if and only if} \quad \lim_{{x \to c^+}} f(x) = L \quad \text{and} \quad \lim_{{x \to c^-}} f(x) = L
   $

### Infinite Limits and Limits at Infinity

1. **Infinite Limits:**
   $
   \lim_{{x \to c}} f(x) = \infty
   $
   This means that $ f(x) $ grows without bound as $ x $ approaches $ c $.

   $
   \lim_{{x \to c}} f(x) = -\infty
   $
   This means that $ f(x) $ decreases without bound as $ x $ approaches $ c $.

2. **Limits at Infinity:**
   $
   \lim_{{x \to \infty}} f(x) = L
   $
   This means that as $ x $ grows without bound, $ f(x) $ approaches $ L $.

   $
   \lim_{{x \to -\infty}} f(x) = L
   $
   This means that as $ x $ decreases without bound, $ f(x) $ approaches $ L $.

### Continuity and Limits

A function $ f $ is continuous at a point $ c $ if:
$
\lim_{{x \to c}} f(x) = f(c)
$
This means the limit of $ f(x) $ as $ x $ approaches $ c $ equals the function value at $ c $.

### Important Limit Laws and Techniques

1. **Direct Substitution:**
   If $ f $ is continuous at $ c $, then $ \lim_{{x \to c}} f(x) = f(c) $.

2. **Factoring:**
   To find a limit, sometimes factoring the expression and canceling common factors can simplify the process.

3. **Rationalization:**
   For limits involving radicals, multiplying by the conjugate can help simplify the expression.

4. **L'HÃ´pital's Rule:**
   If a limit results in an indeterminate form like $ \frac{0}{0} $ or $ \frac{\infty}{\infty} $, then:
   $
   \lim_{{x \to c}} \frac{f(x)}{g(x)} = \lim_{{x \to c}} \frac{f'(x)}{g'(x)},
   $
   provided the limit on the right exists.

Understanding these properties and rules of limits forms the foundation for much of calculus and is essential for analyzing the behavior of functions.

```python
import numpy as np 
import sympy as sym
```

```python
"""
F(x) = (x^3/3)+100*sqrt(abs(x))
"""

s_x = sym.var('x')

s_y = (s_x**3/3)+100* sym.sqrt(sym.Abs(s_x))

c = 3;

target = 5;

"""     
                       Left Hand Side     Right Hand Side
Property: Factorizable lim x-> a c*f(x) = c* lim x->a f(x)
"""

fx = c*s_y # c*f(x)

lim_function = sym.limit(s_y,s_x,target,dir="+-")  # lim x-> target f(x)

RightHand_Side_eq = c*lim_function  # c* (lim x-> target f(x) )


LeftHand_Side_eq = sym.limit(fx,s_x,target,dir="+-")

if(RightHand_Side_eq == LeftHand_Side_eq):
    print("Limits can be Factorised when multiplied by a constant")
    print("LHS lim x->a (c*f(x)) :")
   
    print(LeftHand_Side_eq)

    print("\n")

    print('RHS  c* (lim x->a f(x)):')
    print(RightHand_Side_eq)

```

```python
## Sum Rule :
## lim x->a (f(x)+g(x)) = lim x->a f(x) + lim x->a g(x)

target = 5;

fx = s_x**2 + 3;

gx = sym.Abs(s_x**3)

fx_sum_gx = fx + gx;

lim_fx = sym.limit(fx,s_x,target,dir="+-")

lim_gx = sym.limit(gx,s_x,target,dir="+-")

# lim x->a (f(x)+g(x))
lim_fx_sum_gx = sym.limit(fx_sum_gx,s_x,target,dir="+-")

# lim x->a f(x) + lim x->a g(x)
lim_fx_sum_lim_gx = lim_fx + lim_gx;

if(lim_fx_sum_lim_gx == lim_fx_sum_gx):
  
    print(" lim x->a f(x) + lim x->a g(x):")
   
    print(lim_fx_sum_lim_gx)

    print("\n")

    print('lim x->a (f(x)+g(x))')
    print(lim_fx_sum_gx)
```

```python
## Product Property 


## lim x->a (f(x)g(x)) = lim x->a f(x) * lim x->a g(x)

target = sym.pi

fx = sym.log(sym.sin(sym.sqrt(sym.Abs(s_x))))
gx = sym.exp(sym.sin(sym.sqrt(s_x)))

# Compute the limits
limit_fx = sym.limit(fx, s_x, target, dir="+-")
limit_gx = sym.limit(gx, s_x, target, dir="+-")
fx_gx = fx * gx

# Compute the product of the limits

# lim x->a f(x) * lim x-> g(x)
lim_fx_lim_gx = sym.simplify(limit_fx * limit_gx)

# Compute the limit of the product of the functions

# lim x->a f(x)*g(x)

lim_fx_gx = sym.simplify(sym.limit(fx_gx, s_x, target, dir="+-"))

if(lim_fx_lim_gx==lim_fx_gx):
    print("lim x->a (f(x)g(x)) = lim x->a f(x) * lim x->a g(x)")
    print("\n")
    print("lim x->a f(x)*g(x)")
    print(lim_fx_gx)
    print("\n")
    print(" lim x->a f(x) * lim x->a g(x)")
    print(lim_fx_lim_gx)



```

```python
# lim x->a (f(x)/g(x)) = lim x->a f(x) / lim x->a g(x)
    
fx = sym.log(s_x)+s_x**2;

gx = sym.exp(-s_x)+s_x**3;

hx = s_x**3 + s_x**2 + s_x;

target = sym.pi;


lim_fx_by_gx = sym.simplify(sym.limit((fx/gx),s_x, target,dir="+-"))


lim_fx_by_lim_gx = sym.simplify( 
        (
            sym.limit(fx,s_x,target,dir="+-")/(sym.limit(gx,s_x,target,dir="+-") )
        )  
        
)

target =0;

lim_gx_by_lim_hx = sym.simplify( 
        (
            sym.limit(gx,s_x,target,dir="+-")/(sym.limit(hx,s_x,target,dir="+-") )
        )  
        
)

lim_gx_hx = sym.simplify(sym.limit((gx/hx),s_x, target,dir="+-"))



if(lim_gx_hx == lim_gx_by_lim_hx):
    print("lim x->a (f(x)/g(x)) = lim x->a f(x) / lim x->a g(x)")

sym.plot(fx/gx)
```
