## Excercise 3 (Infinite Limits)



-----
 evaluating infinite limits, which typically involve expressions where $ x $ approaches a value that causes the function's value to grow without bound or approach infinity. Let's analyze the infinite limits for the function $ f(x) = \frac{1}{(x-2)^2} $.

### Infinite Limits Analysis:

#### 1. **Limit as $ x $ approaches 2 from the positive side ($ x \to 2^+ $)**:
$ \lim_{{x \to 2^+}} \frac{1}{(x-2)^2} $

As $ x $ approaches 2 from the right (positive side), $ (x-2)^2 $ approaches 0 from the positive side (since $ (x-2)^2 $ is positive and getting smaller).

$ \lim_{{x \to 2^+}} \frac{1}{(x-2)^2} = +\infty $

This indicates that as $ x $ gets arbitrarily close to 2 from the right, $ \frac{1}{(x-2)^2} $ becomes arbitrarily large (approaches infinity).

#### 2. **Limit as $ x $ approaches 2 from the negative side ($ x \to 2^- $)**:
$ \lim_{{x \to 2^-}} \frac{1}{(x-2)^2} $

As $ x $ approaches 2 from the left (negative side), $ (x-2)^2 $ also approaches 0 from the positive side (since $ (x-2)^2 $ is positive and getting smaller).

$ \lim_{{x \to 2^-}} \frac{1}{(x-2)^2} = +\infty $

Similarly, approaching 2 from the left causes $ \frac{1}{(x-2)^2} $ to become arbitrarily large (approaches infinity).

### Conclusion:
- $ \lim_{{x \to 2^+}} \frac{1}{(x-2)^2} = +\infty $
- $ \lim_{{x \to 2^-}} \frac{1}{(x-2)^2} = +\infty $

In both cases, as $ x $ approaches 2 from either the positive or negative side, the function $ \frac{1}{(x-2)^2} $ grows without bound, approaching infinity. This is characteristic of infinite limits where the function value increases without bound as the independent variable $ x $ approaches a specific value.


-----

```python
import numpy as np;
import sympy as sym;
import matplotlib.pyplot as plt;

```

```python
s_x = sym.symbols('x')

s_y = 1/(s_x-2)**2;

fx = sym.lambdify(s_x,s_y)

xx = np.linspace(0,4,6000)
plt.title("f(x)= 1/(x-2)^2")
plt.plot(xx,fx(xx))
plt.axvline(x=2, color='green', linestyle='--', label='Critical point at x=2')
plt.grid(True)
plt.legend()
plt.ylim(0,100)
plt.show()
```

```python
target = 2;

limit_left = sym.limit(s_y,s_x,target,dir='-')

limit_right = sym.limit(s_y,s_x,target,dir='+')

both_side = sym.limit(s_y,s_x,target,dir='+-')


print(f"Limit as x->2 from the left {limit_left}")
print(f"Limit as x->2 from the Right {limit_right}")
print(f"Both Sided Limits {both_side}")
```
